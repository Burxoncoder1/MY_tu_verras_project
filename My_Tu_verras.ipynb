```python
import pandas as pd
import sqlite3 as sqlite3
import numpy as np
# from my_ds_babel import Replacement
df1 = pd.read_csv("only_wood_customer_us_1.csv")
# print(df1.head())
# print(df1['Gender'].unique())
genders = {'0':'Male','1':'Female','M':'Male','F':'Female'}
df1['Gender'] = df1['Gender'].replace(genders)
df1['FirstName'] = df1['FirstName'].str.replace('\W','')
df1['FirstName'] = df1['FirstName'].str.title()
df1['LastName'] = df1['LastName'].str.replace('\W','')
df1['LastName'] = df1['LastName'].str.title()
df1['UserName'] = df1['UserName'].str.replace('\W','')
df1['UserName'] = df1['UserName'].str.title()
df1['Email'] = df1['Email'].str.lower()
# print(df1['Age'].unique())
df1['City'] = df1['City'].str.replace('_',' ')
df1['City'] = df1['City'].str.replace('-',' ')
df1['City'] = df1['City'].str.title()
df1['Country'] = 'USA'
df1.rename(columns={'Gender':'gender','FirstName':'first_name','LastName':'last_name','UserName':'user_name','Email':'email','Age':'age','City':'city','Country':'country'},inplace=True)
# print(df1.head(10))



df2 = pd.read_csv("only_wood_customer_us_2.csv",sep = ';',header=None, names=['age','city','gender','name','email'])
# print(df2.head(10))
# print(df2['age'][40:100])
df2['age'] = df2['age'].str.replace('\D','')
# print(df2['age'].unique())
df2['city'] = df2['city'].str.replace('_',' ')
df2['city'] = df2['city'].str.replace('-',' ')
df2['city'] = df2['city'].str.title()
# print(df2['city'].unique())
# print(df2['gender'].unique())
genderss = {'0':'Male', '1':'Female', 'F':'Female', 'M':'Male'}
df2['gender'] = df2['gender'].replace(genderss)
# print(df2['gender'].unique())
df2['name'] = df2['name'].str.replace('\W',' ')
df2['name'] = df2['name'].str.title()
df_name = df2['name'].str.split(expand = True)
df_name[2] == None
df_name = df2['name'].str.split(expand=True)
df2['first_name'],df2['last_name'] = df_name[0],df_name[1]
df2.drop('name', axis = 1, inplace = True)
df2['email'] = df2['email'].str.lower()
# print(df2.head(10))



df3 = pd.read_csv("only_wood_customer_us_3.csv",sep = '\t',skiprows=1, names=['gender','name','email','age','city','country'])
# print(df3['gender'].unique())
df3['gender'] = df3['gender'].str.replace('string_','')
df3['gender'] = df3['gender'].str.replace('boolean_','')
df3['gender'] = df3['gender'].str.replace('character_','')
genderss = {'0':'Male', '1':'Female','M':'Male'} 
df3['gender'] = df3['gender'].replace(genderss) 
# print(df3['gender'].unique())
df3['name'] = df3['name'].str.replace('string_','')
df3['name'] = df3['name'].str.replace('\W',' ')
df3['name'] = df3['name'].str.title()
name_d = df3['name'].str.split(expand = True)
name_d[2] == None
name_d = df3['name'].str.split(expand=True)
df3['first_name'],df3['last_name'] = name_d[0],name_d[1]
df3.drop('name', axis = 1, inplace = True)
df3['email'] = df3['email'].str.replace('string_','')
df3['email'] = df3['email'].str.lower()
# print(df3.head(5))
df3['age'] = df3['age'].str.replace('\D','')
# print(df3['age'].unique())
df3['city'] = df3['city'].str.replace('_',' ')
df3['city'] = df3['city'].str.replace('-',' ')
df3['city'] = df3['city'].str.title()
df3['country'] = 'USA'
df3['city'] = df3['city'].str.replace('String ','')
# print(df3.head(10))



df = pd.concat([df1,df2,df3],ignore_index=True)
# print(df.columns)
new_column_names = ['gender', 'first_name', 'last_name', 'email', 'age', 'city','country' ,'user_name']
df.reindex(columns = new_column_names)
print(df.head(30))
```
